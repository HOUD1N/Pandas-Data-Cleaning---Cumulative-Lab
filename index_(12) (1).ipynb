```python
from itertools import combinations

# Create a dataframe to store the co-occurrence counts
co_occurrence_df = pd.DataFrame(0, index=powers_df.columns, columns=powers_df.columns)

# Iterate over each hero and update the co-occurrence dataframe
for _, row in heroes_and_powers_df.iterrows():
    powers = row[powers_df.columns].fillna(False)  # Convert NaNs to False
    active_powers = powers[powers].index.tolist()  # Only consider powers that the hero has
    for power1, power2 in combinations(active_powers, 2):
        co_occurrence_df.loc[power1, power2] += 1
        co_occurrence_df.loc[power2, power1] += 1  # Since co-occurrence is symmetric

# Display the co-occurrence dataframe
print(co_occurrence_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Flatten the co-occurrence matrix and sort by frequency
co_occurrence_flat = co_occurrence_df.unstack().sort_values(ascending=False)

# Drop the diagonal (self-co-occurrence)
co_occurrence_flat = co_occurrence_flat[co_occurrence_flat.index.get_level_values(0) != co_occurrence_flat.index.get_level_values(1)]

# Display the top 10 most common co-occurrences
print(co_occurrence_flat.head(10))

# Plot a heatmap of the co-occurrence matrix
plt.figure(figsize=(12, 10))
sns.heatmap(co_occurrence_df, cmap="YlGnBu", annot=False)
plt.title("Superpower Co-Occurrence Heatmap")
plt.show()
```

Top Pairs: The top pairs like "Super Strength and Durability" or "Super Strength and Stamina" suggest that heroes with these abilities are common, possibly because they form a core set of powers for many superheroes.
Visualization: The heatmap helps in quickly spotting clusters or specific areas where certain powers tend to co-occur frequently, providing insights into common superpower combinations.
